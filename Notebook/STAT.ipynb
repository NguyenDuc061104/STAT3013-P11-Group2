# **Calculate basic statistics**

```python
# Calculate basic statistics

#Calculate the average of all values â€‹â€‹in the TotalCharges column.
mean_total_charges = df['TotalCharges'].mean()

#Calculate the median, which is the middle value of TotalCharges when sorting values â€‹â€‹in ascending order.
median_total_charges = df['TotalCharges'].median()

#Find trend â€“ the value that appears most often in the TotalCharges column.
#[0] is used to get the first value of the mode, because mode() returns a Series (if there are multiple modes, only get the first value).
mode_total_charges = df['TotalCharges'].mode()[0]

#Get the maximum value and data['TotalCharges'].min() get the smallest value in the TotalCharges column.
range_total_charges = df['TotalCharges'].max() - df['TotalCharges'].min()

#Calculate the variance of TotalCharges, which shows how dispersed the values â€‹â€‹are from the mean.
#The higher the variance, the more scattered the values â€‹â€‹are around the mean.
variance_total_charges = df['TotalCharges'].var()

# Calculate the standard deviation, which is the square root of the variance.
std_dev_total_charges = df['TotalCharges'].std()

#Print the result
print("Mean:", mean_total_charges)
print("Median:", median_total_charges)
print("Mode:", mode_total_charges)
print("Range:", range_total_charges)
print("Variance:", variance_total_charges)
print("Standard Deviation:", std_dev_total_charges)
```

# **Logistic Regression Model**

```python
# Import necessary libraries

#This line imports the LogisticRegression class from the sklearn.linear_model library.
#LogisticRegression is a machine learning model used for binary or multiclass classification, using logistic regression to predict the probabilities of classes.
from sklearn.linear_model import LogisticRegression

#Split the data into training set and test set.
from sklearn.model_selection import train_test_split

#Evaluate model accuracy.
from sklearn.metrics import accuracy_score
```

```python
# Step 1: Initialize and train the Logistic Regression model

#Initialize the Logistic Regression model
lr_model = LogisticRegression()
#Train the model
lr_model.fit(X_train,y_train)
```

```python
# Step 2: Calculate the model's accuracy

#lr_model.score(X_test, y_test) calculates the accuracy of the lr_model on the test data (X_test and y_test)
accuracy_lr = lr_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy_lr)
```

```python
# Step 3: Predict and create assessment report

# Import classification_report
from sklearn.metrics import classification_report

#lr_model.predict(X_test) uses the trained lr_model to predict the label (y) for the data in X_test.
lr_pred= lr_model.predict(X_test)

#classification_report(y_test, lr_pred) generates an evaluation report of the model's performance metrics, comparing the actual label (y_test) with the predicted label (lr_pred).
report = classification_report(y_test,lr_pred)
print(report)
```

```python
# Step 4: Draw the Confusion Matrix

from sklearn.metrics import confusion_matrix # Import confusion_matrix

#plt.figure(figsize=(4,3)) sets the size of the chart (4 units wide and 3 units high).
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, lr_pred),
                annot=True,fmt = "d",linecolor="k",linewidths=3)

#plt.title(...) adds the title "LOGISTIC REGRESSION CONFUSION MATRIX" with a font size of 14 to describe the chart type.
plt.title("LOGISTIC REGRESSION CONFUSION MATRIX",fontsize=14)
plt.show()
```

```python
# Import the roc_curve function from sklearn.metrics
from sklearn.metrics import roc_curve

#lr_model.predict_proba(X_test) returns a 2D array, where each line represents the probability of each class (0 and 1) for each data point in the test set.
#[:, 1] extracts the probability for the positive class (usually class 1). These probabilities are needed to draw the ROC curve.
y_pred_prob = lr_model.predict_proba(X_test)[:,1]

#fpr: False positive rate at different threshold values.
#tpr: True positive rate at different threshold values.
#thresholds: Threshold values â€‹â€‹used to classify a data point as positive or negative.
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

#Draw a diagonal line from (0,0) to (1,1). This line represents the performance of a stochastic model (prediction like a guess), serving as the baseline.
plt.plot([0, 1], [0, 1], 'k--' )

#Draw the actual ROC curve for the logistic regression model using fpr and tpr values.
plt.plot(fpr, tpr, label='Logistic Regression',color = "r")

#Label the x-axis "False Positive Rate", label the y-axis "True Positive Rate", and the title "Logistic Regression ROC Curve".
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve',fontsize=16)

plt.show();
```
