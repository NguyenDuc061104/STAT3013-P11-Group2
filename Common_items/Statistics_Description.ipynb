# **Calculate basic statistics**

```python
# Calculate basic statistics

#Calculate the average of all values ​​in the TotalCharges column.
mean_total_charges = df['TotalCharges'].mean()

#Calculate the median, which is the middle value of TotalCharges when sorting values ​​in ascending order.
median_total_charges = df['TotalCharges'].median()

#Find trend – the value that appears most often in the TotalCharges column.
#[0] is used to get the first value of the mode, because mode() returns a Series (if there are multiple modes, only get the first value).
mode_total_charges = df['TotalCharges'].mode()[0]

#Get the maximum value and data['TotalCharges'].min() get the smallest value in the TotalCharges column.
range_total_charges = df['TotalCharges'].max() - df['TotalCharges'].min()

#Calculate the variance of TotalCharges, which shows how dispersed the values ​​are from the mean.
#The higher the variance, the more scattered the values ​​are around the mean.
variance_total_charges = df['TotalCharges'].var()

# Calculate the standard deviation, which is the square root of the variance.
std_dev_total_charges = df['TotalCharges'].std()

#Print the result
print("Mean:", mean_total_charges)
print("Median:", median_total_charges)
print("Mode:", mode_total_charges)
print("Range:", range_total_charges)
print("Variance:", variance_total_charges)
print("Standard Deviation:", std_dev_total_charges)
```

# **TotalCharges Distribution Chart**

```python
#set the size of the chart with a width of 10 units and a height of 6 units.
plt.figure(figsize=(10, 6))

#sns.histplot() is a Seaborn function used to draw a distribution chart (histogram) for numerical data.
#df['TotalCharges'] is the data to plot, here is the TotalCharges column in DataFrame df.
#bins=30 divides the data into 30 groups (bins), adjusting the smoothness of the chart.
#kde=True adds a density line (Kernel Density Estimation) to the chart, helping to more clearly see the shape of the data's distribution.
sns.histplot(df['TotalCharges'], bins=30, kde=True)

#plt.title('TotalCharges Distribution Chart') gives the title of the chart “TotalCharges Distribution Chart”.
plt.title('TotalCharges Distribution Chart')

#plt.xlabel('TotalCharges') labels the x-axis "TotalCharges", representing the cost data axis.
plt.xlabel('TotalCharges')

#plt.ylabel('Number of customers') labels the y-axis "Number of customers", representing the number of customers at each TotalCharges level.
plt.ylabel('Number of customers')

#plt.show() displays the graph on the screen.
plt.show()
```

# **Box Plot of TotalCharges and MonthlyCharges**

```python
# Plot for TotalCharges
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['TotalCharges']])
plt.title('Box plot of TotalCharges')
plt.ylabel('Cost Value')
plt.show()
```

```python
# Plot for MonthlyCharges
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['MonthlyCharges']])
plt.title('Box plot of MonthlyCharges')
plt.ylabel('Cost Value')
plt.show()
```

# **Scatter plot between MonthlyCharges and TotalCharges**

```python
plt.figure(figsize=(10, 6))
#alpha=0.5 sets the transparency for the data points, with alpha at 0.5, helping to reduce overlap if there are multiple points in the same location.
sns.scatterplot(data=df, x='MonthlyCharges', y='TotalCharges', alpha=0.5)
plt.title('Scatter plot of TotalCharges and MonthlyCharges')
plt.xlabel('MonthlyCharges')
plt.ylabel('TotalCharges')
plt.show()
```
