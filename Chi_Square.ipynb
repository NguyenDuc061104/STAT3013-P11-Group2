```python
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

def perform_chi_square_test(df, col1, col2):
    # Create cross-frequency table
    contingency_table = pd.crosstab(df['TotalCharges'], df['MonthlyCharges'])

    # Perform Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Conclusion is statistically significant at 5% significance level
    significant = p < 0.05  # True nếu có ý nghĩa thống kê

    return chi2, p, significant

# Call function with data
chi2, p, significant = perform_chi_square_test(df, 'TotalCharges', 'MonthlyCharges')

# Print results
print("Chi-square:", chi2)
print("P-value:", p)
print("Significant:", significant)
```
